## Q1. What is the difference between a function and a method in Python?
In Python, a function is a block of reusable code that is defined using the def keyword and can be called independently, whereas a method is a function that is associated with an object (usually a class instance). Methods are called on objects and can access and modify the object’s data, while functions operate independently unless passed an object.

### Example:
####Function
    def greet(name):
      return f"Hello, {name}!"

    print(greet("Alice"))

#### Method
    name = "Alice"
    print(name.upper())  # .upper() is a method of string objects


## Q2. Explain the concept of function arguments and parameters in Python.
Parameters are variables listed inside the parentheses in the function definition. Arguments are the values passed to the function when it is called. Parameters receive the arguments so the function can operate on them. Python also supports default, keyword, and variable-length arguments for flexibility.

### Example:


    def add(a, b):  # 'a' and 'b' are parameters
      return a + b

    print(add(3, 5))  # 3 and 5 are arguments


## Q3. What are the different ways to define and call a function in Python?
Functions in Python can be defined using def for standard functions or lambda for anonymous functions. They can be called by their name followed by parentheses containing any required arguments. Functions can also be stored in variables or passed as arguments.

###Example:

    # Regular function
    def square(x):
        return x * x

    print(square(4))

    # Lambda function
    double = lambda x: x * 2
    print(double(5))

    # Function as argument
    def apply(func, value):
        return func(value)

    print(apply(square, 6))


## Q4. What is the purpose of the return statement in a Python function?
The return statement is used to send a result back from a function to the caller. It ends the function’s execution and provides the computed value. Without return, a function will return None by default.

### Example:
    def multiply(a, b):
        return a * b

    result = multiply(3, 4)
    print(result)  # Output: 12


## Q5. What are iterators in Python and how do they differ from iterables?
An iterable is any object that can return an iterator, typically using iter(). Lists, tuples, strings, etc., are iterables. An iterator is an object that keeps state and produces the next value when next() is called. The key difference is that iterables can be looped over, but iterators generate items one at a time.

###Example:
    # Iterable
    nums = [1, 2, 3]

    # Iterator
    it = iter(nums)
    print(next(it))  # 1
    print(next(it))  # 2


## Q6. Explain the concept of generators in Python and how they are defined.
A generator is a special type of iterator that yields values one at a time using the yield keyword. They are defined like regular functions but use yield instead of return. They are memory-efficient because they generate items on the fly.

### Example:
    def count_up_to(n):
    i = 1
    while i <= n:
        yield i
        i += 1

    for num in count_up_to(3):
        print(num)


## Q7. What are the advantages of using generators over regular functions?
Generators are more memory-efficient because they yield values one at a time and don’t store the entire output in memory. They are ideal for working with large data sets, infinite sequences, or when results are needed one-by-one instead of all at once.

###Example:
    def generate_numbers():
        for i in range(1000000):
            yield i

    # This doesn't load all million numbers into memory at once
    for num in generate_numbers():
        if num > 5:
            break
        print(num)


## Q8. What is a lambda function in Python and when is it typically used?
A lambda function is an anonymous, single-expression function defined with the lambda keyword. It’s commonly used for short, throwaway functions, especially as arguments to higher-order functions like map(), filter(), and sorted().

###Example:
    # Lambda to double a number
    double = lambda x: x * 2
    print(double(4))  # Output: 8


## Q9. Explain the purpose and usage of the map() function in Python.
The map() function applies a given function to all items in an iterable (like a list) and returns a map object, which can be converted to a list. It's useful for transforming data without explicit loops.

###Example:
    nums = [1, 2, 3, 4]
    squares = list(map(lambda x: x**2, nums))
    print(squares)  # Output: [1, 4, 9, 16]


## Q10. What is the difference between map(), reduce(), and filter() functions in Python?
- map() applies a function to each item in an iterable and returns a new iterable.

- filter() selects items from an iterable based on a condition (function returns True).

- reduce() (from functools) applies a rolling computation to reduce the iterable to a single value.

### Example:
    from functools import reduce

    nums = [1, 2, 3, 4]

    # map
    squared = list(map(lambda x: x**2, nums))  # [1, 4, 9, 16]

    # filter
    evens = list(filter(lambda x: x % 2 == 0, nums))  # [2, 4]

    # reduce
    product = reduce(lambda x, y: x * y, nums)  # 24

    print(squared, evens, product)

### Q1.Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

```python
def sum_even(l):
  sum_even=0
  for i in l:
    if i%2==0:
      sum_even+=i
  return sum_even

l=list(map(int,input("Enter the list elements seperated by space: ").split()))
print(sum_even(l))
```

### Q4. Write a Python function that checks if a given number is prime or not from 1 to 200.


```python
def is_prime(n):
    if n <= 1:
        return False  # 0 and 1 are not prime
    for i in range(2, int(n**0.5) + 1):  # Only check up to √n
        if n % i == 0:
            return False
    return True

# Test: Check primes from 1 to 200
for num in range(1, 201):
    if is_prime(num):
        print(num, "is a prime number")
```

### Q5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.


```python
def fibonacci(n):
  fibonacci_seq=[]
  a,b=0,1
  fibonacci_seq.append(a)
  for i in range(n-1):
    a,b= b,a+b
    fibonacci_seq.append(a)
  return fibonacci_seq

n=int(input("Enter the number of terms: "))
print(fibonacci(n))
```

### Q6. 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.


```python
def power_of_2(n):
  for i in range(n):
    yield 2**i

n=int(input("Enter the exponent: "))
for i in power_of_2(n):
  print(i)
```

### Q8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.


```python

user_input = input("Enter tuples (e.g. (1,3,5) (2,1) (4,2,6,7) (5,0,9)) with spaces : ")

# Convert input string to list of tuples
data = [tuple(map(int, item.strip('()').split(','))) for item in user_input.split()]

# Sort based on second element using lambda
# Make sure each tuple has at least 2 elements to avoid IndexError
sorted_data = sorted(data, key=lambda x: x[1] if len(x) > 1 else float('inf'))

print("Sorted list:", sorted_data)
```

### Q9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

### Q10. Create a Python program that uses `filter()` to remove all the vowels from a given string.


```python
string = input("Enter a string: ")
vowels = ['a', 'e', 'i', 'o', 'u', 'A', 'E', 'I', 'O', 'U']
filtered_string = ''.join(filter(lambda x: x not in vowels, string))
print("Filtered string:", filtered_string)
```

##Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the product of the price per item and the quantity. The product should be increased by 10,- € if the value of the order is smaller than 100,00 €. Write a Python program using lambda and map

```python
# Given data
orders = [
    [34587, "Learning Python, Mark Lutz", 4, 40.95],
    [98762, "Programming Python, Mark Lutz", 5, 56.80],
    [77226, "Head First Python, Paul Barry", 3, 32.95],
    [88112, "Einführung in Python3, Bernd Klein", 3, 24.99]
]

# Using map and lambda
invoice = list(map(lambda order: (order[0], round(order[2] * order[3] + (10 if order[2] * order[3] < 100 else 0), 2)), orders))

print(invoice)
```
